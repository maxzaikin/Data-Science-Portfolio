###  Анализ влияния отзывов пользователей и критиков на продажи внутри ТОП платформ .

Посмотрите, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. Постройте диаграмму рассеяния и посчитайте корреляцию между отзывами и продажами. Сформулируйте выводы.

```python
legend_list = []
top_sales_platforms = [['PS3'],['X360'],['PS4'],['3DS']]

pierson_koef = pd.DataFrame(columns=['platform','wsales-crit_rating','wsales-usr_rating'])


for platform in top_sales_platforms:
   
    top_sales_platform_df = data.query('platform == @platform & year_of_release >=2013')
    
    ax = top_sales_platform_df.plot(x='user_score',
                               xlabel='Оценки пользователей',
                               y='world_sales',
                               ylabel='Кол-во продаж (млн. шт.)',
                               kind='scatter', 
                               title=f'Диаграма рассеяния между оценками пользователей и продажами {platform}',
                               alpha=0.5,
                               figsize=(19, 4),
                               fontsize=12,
                               c='user_score',
                               colormap='viridis') 
    
    ax = top_sales_platform_df.plot(x='user_score',
                                    xlabel='Оценки пользователей',
                                    y='world_sales',
                                    ylabel='Кол-во продаж (млн. шт.)',
                                    kind='hexbin', 
                                    gridsize=10, 
                                    figsize=(5, 4), 
                                    sharex=False, 
                                    grid=True) 
    
    ax = top_sales_platform_df.plot(x='critic_score',
                               xlabel='Оценки критиков',
                               y='world_sales',
                               ylabel='Кол-во продаж (млн. шт.)',
                               kind='scatter', 
                               title=f'Диаграма рассеяния между оценками критиков и продажами {platform}',
                               alpha=0.5,
                               figsize=(19, 4),
                               c='critic_score',
                               colormap='viridis')    
    
    
    ax = top_sales_platform_df.plot(x='critic_score',
                                    xlabel='Оценки критиков',
                                    y='world_sales',
                                    ylabel='Кол-во продаж (млн. шт.)',
                                    kind='hexbin', 
                                    gridsize=10, 
                                    figsize=(5, 4), 
                                    sharex=False, 
                                    grid=True) 
    
    pierson_koef = pierson_koef.append({'platform':platform,
                        'wsales-crit_rating':top_sales_platform_df['world_sales'].corr(top_sales_platform_df['critic_score']).round(2),
                        'wsales-usr_rating':top_sales_platform_df['world_sales'].corr(top_sales_platform_df['user_score']).round(2)},
                        ignore_index=True)
    

plt.show()   

print('Расчитаем колереляцию Пирсона между продажами и оценками')
pierson_koef
```

На основе полученых данных мы видим что в большинстве случаев кореляция между продажами и оценками практически отсутствует. 
    Здесь стоит отметить однако что коэфициент кореляции между продажами флагманских консолей оценками пользователей немного выше (колонка wsales-usr_rating) однако и в пике не достигает 0.5, что так же говорит об отсутсвии ярко выраженной зависимости, и скорее всего зависимость носит спонтанный характер.
