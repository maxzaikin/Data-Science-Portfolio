Заказчик: компания занимающаяся пассажирскими перевозками собрала исторические данные о заказах такси в аэропортах, чтобы привлекать больше водителей в период пиковой нагрузки.

Задача: построить модель для прогноза количества заказов такси на следующий час.

Условия: Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.



### Functions

```python
def histnbox_plot_for_feature(df, feature_name):
    '''
    Функция вывода гистограмм и ящика с усами для количесвтенных данных.
    Функция сама определяет все колличественные серии, а затем строит для них графики
    Args:
        df: исходный датасет, по которым необходимо построить графики.
        feature_name: Признак для которго необходимо построить графики
    Return:
        Гистрограмма и боксплот для конкретного признака
    '''
    sns.set_style('whitegrid')
    title         = 'Гистограмма и ящик с усами для: ' + feature_name    
    f, ax         = plt.subplots(1,2, figsize=(16, 8))
    
    f.tight_layout()
    f.set_figheight(10)
    f.set_figwidth(20)
    plt.rcParams.update({'font.size': 10}) 
    
    sns.histplot(data = df[feature_name],                     
                kde  = True, 
                bins = 50, 
                ax   = ax[0])  
        
    sns.boxplot(y = df[feature_name], 
                ax   = ax[1],
                showmeans=True,
                meanprops={"marker":"s","markerfacecolor":"white", "markeredgecolor":"blue", "markersize":"5"})   
    ###################
    #   Гистограмма
    # Заголовок
    ax[0].set_title('Серия: ' + feature_name + ' . Тип: ' + str(df.dtypes[feature_name]))
    # Подпись по оси x
    ax[0].set_xlabel('Параметр: ' + feature_name)
    # Подпись по оси y
    ax[0].set_ylabel('Количество')
        
    ####################
    #    Ящик с усами
    # Заголовок
    ax[1].set_title('Серия: ' + feature_name)
        
    # Подпись по оси x
    ax[1].set_xlabel('Параметр: ' + feature_name)    
            
    plt.suptitle(title, fontsize=20, y=1.01)
    plt.show()
    
def plot_timeseries(df, title):
    '''
    Функция отрисовки временного ряда с использованием библиотеки Seaborn
    Args:
        df: Датасет в виде временного ряда
        title: заголовок графика
    Return:
        График временного ряда
    '''
    sns.set_style('whitegrid')
    plt.figure(figsize=(12, 6))
    sns.lineplot(x=df.index, y="num_orders",data=df)
    plt.xticks(rotation=45)

    plt.title(title)
    plt.xlabel('Дата')
    plt.ylabel('Количество заказов')
    plt.show()    

def plot_timeseries_decompose(df, resample_interval, title):
    '''
    Функция отрисовки графиков декомпозиции временного ряда на основе поданного на вход датасета
    Args:
        df: Датасет
        resample_interval: Интервал группировки в различных комбинациях. Примеры:
                        'H' - по часам
                        'D' - по дням
                        'W' - по неделям
                        'M' - по месяцам
                        'Y' - по годам
        title: Заголовок графика
        
    Return:
        На выходе три графика: Тренд, Сезонность, Шумы
    '''
    rotation = 90
    labelfontsize = 10
    tickfontsize = 8
    formatter = '%Y-%m-%d (%a)'# Формат даты с названием дня недели
    labelpadsize = 40
    figsize = (15, 4)
    
    df_resampled = df.resample(resample_interval).sum()
    decomposed = seasonal_decompose(df_resampled)
    
    fig, axes = plt.subplots(nrows=1, ncols=1, figsize=figsize)

    axes.plot(decomposed.trend)
    axes.set_title('Тренд/Trend ' + title)
    axes.tick_params(axis='x', rotation=rotation, labelsize=tickfontsize, pad=labelpadsize)
    axes.xaxis.set_major_locator(mdates.DayLocator())
    axes.xaxis.set_major_formatter(mdates.DateFormatter(formatter))  
    axes.set_xlabel('Дата/Date', fontsize=10) 
    plt.show()

    fig, axes = plt.subplots(nrows=1, ncols=1, figsize=figsize)

    axes.plot(decomposed.seasonal)
    axes.set_title('Сезонность/Seasonality ' + title)
    axes.tick_params(axis='x', rotation=rotation, labelsize=tickfontsize, pad=labelpadsize)
    axes.xaxis.set_major_locator(mdates.DayLocator())
    axes.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d (%a)'))
    axes.set_xlabel('Дата/Date', fontsize=10) 
    plt.show()
        
    fig, axes = plt.subplots(nrows=1, ncols=1, figsize=figsize)
                             
    axes.plot(decomposed.resid)
    axes.set_title('Шумы/Residuals ' + title)
    axes.tick_params(axis='x', rotation=rotation, labelsize=tickfontsize, pad=labelpadsize)
    axes.xaxis.set_major_locator(mdates.DayLocator())
    axes.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d (%a)'))
    axes.set_xlabel('Дата/Date', fontsize=10) 
    plt.show()
    
def resid_analysis(df):
    rotation = 90
    labelfontsize = 10
    tickfontsize = 8
    formatter = '%Y-%m-%d (%a)'# Формат даты с названием дня недели
    labelpadsize = 40
    figsize = (15, 4)

    df_resampled = df.resample('D').sum()
    decomposed = seasonal_decompose(df_resampled)

    rolling_mean = decomposed.resid.rolling(window=7).mean()

    fig, axes = plt.subplots(nrows=1, ncols=1, figsize=figsize)

    # Построение графика остатков
    axes.plot(decomposed.resid, label='Остатки/Residuals')
    axes.set_title('Анализ Шумов / Residuals')
    axes.tick_params(axis='x', rotation=rotation, labelsize=tickfontsize, pad=labelpadsize)
    axes.xaxis.set_major_locator(mdates.MonthLocator())
    axes.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    axes.set_xlabel('Дата/Date', fontsize=10)

    # Добавление графика скользящего стандартного среднего
    axes.plot(rolling_mean, label='Скользящее стандартное среднее/Moving Average')
    axes.legend()

    plt.show()
    
 
```

#### SARIMA

p-lags (p):
- Порядок авторегрессии (AR): Определяет, сколько прошлых значений ряда используется для прогнозирования текущего значения.
- Пример: AR(p) модель использует p прошлых значений.
- Влияние: Большее значение p указывает на более сильную зависимость текущего значения от прошлых значений.

d (порядок интегрирования): 
- Определяет, сколько раз нужно продифференцировать ряд, чтобы сделать его стационарным.

q-lags (q):
- Порядок скользящего среднего (MA): Определяет, сколько прошлых ошибок прогноза (разница между фактическими и прогнозируемыми значениями) используется для корректировки текущего прогноза.
- Пример: MA(q) модель использует q прошлых ошибок.
- Влияние: Большее значение q указывает на более сильную реакцию модели на прошлые ошибки прогноза.

P (сезонный порядок авторегрессии): 
- Аналогично p, но для сезонных лагов.

D (сезонный порядок интегрирования):
- Аналогично d, но для сезонного дифференцирования.

Q (сезонный порядок скользящего среднего): 
- Аналогично q, но для сезонных лагов ошибок прогноза.

m (сезонный период): 
- Количество наблюдений в каждом сезоне (например, 12 для месячных данных)

```python
# Начальное значение параметра d
best_d = 0
regression='ctt' # применяем ctt так как ттенд квадритичный

# Начальное значение p_value на нулевом d
best_p_value = adfuller(modeling_data['num_orders'], regression=regression)[1]

# Проверяем стационарность исходного ряда
print(f"Параметр d = {best_d}, p-value: {best_p_value}")

# Пока ряд не станет стационарным, увеличиваем d и повторяем тест
d = 1
while best_p_value >= 0.05:
    ts_diff = modeling_data['num_orders'].diff(periods=d)  # Применяем разность с учетом текущего d
    ts_diff.dropna(inplace=True)  # Удаляем пропущенные значения
    
    # Тестируем стационарность ряда после применения разности
    p_value= adfuller(ts_diff, regression=regression)[1]
    print(f"Параметр d = {d}, p-value: {p_value}")
    
    # Если p-value лучше (меньше) AIC, обновляем лучшее значение d и p-value
    if p_value < best_p_value:
        best_p_value = p_value
        best_d = d
    
    # Если p-value становится хуже (больше) AIC, завершаем цикл
    else:
        break
    
    d += 1

# Когда ряд станет стационарным, выводим оптимальное значение параметра d
print(f"Оптимальное значение параметра d: {best_d}")

# Заглушка, в связи с тем что подбор количества дифференцирований не требуется, так как дифференцирование было проведено ранее
#best_d = 0 - Убираем заглушку
```

Общая информация:
- SARIMAX использует метод оптимизации, называемый L-BFGS-B, для поиска наилучших значений параметров модели, которые максимизируют функцию правдоподобия.
  - "f" означает значение целевой функции, которую алгоритм пытается оптимизировать. В данном случае, это значение функции правдоподобия, которое должно быть максимально возможным.
  - "|proj g|" - это норма проекции градиента. Градиент показывает направление наискорейшего увеличения функции, а его проекция это часть градиента, которая учитывает ограничения на параметры модели. 
  
- Тип ковариционной матрицы - opg (outer product of gradients). Метод оценки ковариационной матрицы коэффициентов модели. Он основан на внешнем произведении градиентов функции правдоподобия.
- coef: Оценка коэффициента модели.
- std err: Стандартная ошибка коэффициента, показывает неопределенность оценки.
- z: z-статистика, используется для проверки гипотезы о том, что коэффициент равен нулю.
- P>|z|: p-значение, показывает вероятность получения наблюдаемого значения z-статистики.
- [0.025 0.975]: 95% доверительный интервал для коэффициента. Это диапазон значений, в котором с 95% вероятностью находится истинный коэффициент.
  - Чем уже интервал, тем точнее оценка коэффициента.
  - Если интервал не включает ноль, это означает, что коэффициент статистически значим, и мы можем быть уверены, что он оказывает влияние на зависимую переменную.
  - Если интервал включает ноль, это означает, что коэффициент может быть не статистически значим, и мы не можем быть уверены в его влиянии на зависимую переменную.

- AIC(Akaike Information Criterion): 
  - AIC = 2k - 2ln(L), где k - количество параметров в модели, L - максимальное значение функции правдоподобия.
  - Интерпретация: AIC оценивает качество модели, учитывая компромисс между точностью и сложностью. Модель с меньшим значением AIC считается предпочтительнее, так как она достигает хорошей точности с меньшим количеством параметров.
  - Комментарий: AIC полезен для сравнения моделей с одинаковой зависимой переменной, но разным количеством параметров.
    40824.683 (чем ниже, тем лучше)
    
- BIC(Bayesian Information Criterion): 
  - BIC = kln(n) - 2ln(L), где k - количество параметров в модели, n - размер выборки, L - максимальное значение функции правдоподобия.
  - Интерпретация: BIC похож на AIC, но он более строго наказывает модели с большим количеством параметров, особенно при большом размере выборки.
  - Комментарий: BIC полезен для выбора модели, которая наиболее вероятно является истинной моделью, генерирующей данные.
  40863.039 (чем ниже, тем лучше)
  
- HQIC(Hannan-Quinn Information Criterion): 40838.211 (чем ниже, тем лучше)
  - HQIC = 2kln(ln(n)) - 2ln(L), где k - количество параметров в модели, n - размер выборки, L - максимальное значение функции правдоподобия.
  - Интерпретация: HQIC похож на AIC и BIC, но он использует другую функцию штрафа для количества параметров. Он менее строг, чем BIC, но более строг, чем AIC.
  - Комментарий: HQIC является альтернативой AIC и BIC и может быть полезен в случаях, когда AIC и BIC дают противоречивые результаты.
  
Сравнение AIC, BIC и HQIC:
- AIC: менее строг к количеству параметров, поэтому может выбрать более сложную модель, даже если она не дает существенного улучшения точности.
- BIC: более строг к количеству параметров и, следовательно, может выбрать более простую модель, даже если она имеет немного худшую точность.
- HQIC: находится между AIC и BIC по строгости к количеству параметров.
  
  
В нашем случае, AIC, BIC и HQIC имеют близкие значения, что говорит о том, что выбранная модель SARIMAX является разумным выбором.

*z-статистика*
z-статистика интерпретируется в контексте стандартной ошибки и показывает, насколько оценка коэффициента модели отличается от нуля в единицах стандартной ошибки
- z = (оценка коэффициента - 0) / стандартная ошибка
  - Большое абсолютное значение z: указывает на то, что оценка коэффициента значительно отличается от нуля, и, следовательно, коэффициент вероятно имеет значимое влияние на зависимую переменную.
  - Маленькое абсолютное значение z: указывает на то, что оценка коэффициента близка к нулю, и коэффициент может не иметь значимого влияния на зависимую переменную.

*P>|z|*
p-значение - это вероятность получения наблюдаемого значения z-статистики (или более экстремального), если истинный коэффициент равен нулю (нулевая гипотеза).

- Низкое p-значение (обычно < 0.05): указывает на то, что маловероятно получить наблюдаемое значение z-статистики, если истинный коэффициент равен нулю. Следовательно, мы отвергаем нулевую гипотезу и заключаем, что коэффициент статистически значим и оказывает влияние на зависимую переменную.

- Высокое p-значение (обычно > 0.05): указывает на то, что вполне вероятно получить наблюдаемое значение z-статистики, даже если истинный коэффициент равен нулю. Следовательно, мы не можем отвергнуть нулевую гипотезу и заключаем, что нет достаточных доказательств того, что коэффициент оказывает значимое влияние на зависимую переменную.
    
    - ar.L1: z = 20.060, P>|z| = 0.000. Это означает, что оценка коэффициента ar.L1 значительно отличается от нуля (20.06 стандартных ошибок), и вероятность получить такое значение z, если истинный коэффициент равен нулю, практически равна нулю. Следовательно, мы заключаем, что коэффициент ar.L1 статистически значим.
    - ma.S.L48: z = 7.791, P>|z| = 0.000. Аналогично, оценка коэффициента ma.S.L48 также значительно отличается от нуля и является статистически значимой.

Модели для поиска улучшения были отключены так как после увеличения колличества лагов ядро стало падать. Оставляю код закоментированым для истории

```python
# Создаем объект TimeSeriesSplit с 5 разделениями, где каждая часть будет использоваться в качестве валидационной выборки, 
# а остальные части - как обучающая выборка.
ts_cv = TimeSeriesSplit(n_splits=5)
scores = []

# Пройдем в цикле по каждому разделению данных    
for _, (train_idx, valid_idx) in enumerate(ts_cv.split(modeling_data['num_orders'])):
    # train_idx: Индексы обучающей выборки для текущего разделения.
    # valid_idx: Индексы валидационной выборки для текущего разделения.
    
    # Извлечем обучающую и валидационную выборки из данных modeling_data['num_orders'].
    # В этом случае, X_train_/val и y_train_/val одинаковы, так как SARIMAX использует саму историю ряда для обучения и прогнозирования.
    X_train_, y_train_ = modeling_data['num_orders'].iloc[train_idx], modeling_data['num_orders'].iloc[train_idx]
    X_val_, y_val = modeling_data['num_orders'].iloc[valid_idx], modeling_data['num_orders'].iloc[valid_idx]

    # Определим начальный и конечный индексы для прогнозирования на валидационной выборке.
    # Начальным индексом будет элемент идущий сразу после последнего элемента обучающей выборки. 
    start = len(X_train_) # Так как индексы начинаются с 0, индекс равен len(X_train_).
    
    #Количество элементов для прогнозирования, то есть end, это длина валидационной выборки, len(X_val_).
    end = len(X_train_) + len(X_val_) - 1 # так как первый элемент вал выборки уже учтен в start, необходимо вычесть 1, чтобы не выйти за пределы выборки.
    
    # Создаеv модель SARIMAX с порядками (best_p, best_d, best_q), которые мы определили ранее.
    model_sarima = SARIMAX(X_train_, order=(best_p,best_d,best_q))

    results_sarima = model_sarima.fit()
    pred_sarima = results_sarima.predict(start, end)
    
    # Делает прогноз на валидационной выборке, и добавим его в массив прогнозов
    scores.append(np.sqrt(mean_squared_error(y_pred=pred_sarima, y_true=y_val)))

print(f'RMSE SARIMA = {round(np.mean(scores), 4)}')
```
