<div>
My client is mobile operator who has found that many customers are using outdated tariffs. They requested to build a system that can analyze customer behavior and suggest one of 2 new tariff to their customers: "Smart" or "Ultra".

Build a model with the highest possible accuracy value (>=0.75) that can analyze customer behavior and suggest a new tariff: "Smart" or "Ultra".

Each object in the dataset is information about the behavior of one user per month. It is known that:

---  ------                --------    --------------------------------------     ------------ 
    - calls    : number of calls

```python
def error_count(answers, predictions):
    """
    Calculates the number of incorrect predictions.

    Args:
        answers (list): A list of true labels.
        predictions (list): A list of predicted labels.

    Returns:
        int: The number of incorrect predictions.
    """
    counter = 0
    for i in range(len(answers)):
        if (answers[i] != predictions[i]):
            counter +=1
    return counter
```

üìå –†–∞–∑–¥–µ–ª–∏–º –∏—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –Ω–∞ –æ–±—É—á–∞—é—â—É—é[TRAIN], –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é[VALID] –∏ —Ç–µ—Å—Ç–æ–≤—É—é[TEST] –≤—ã–±–æ—Ä–∫–∏.

    TRAIN - —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–∞—è –∏–ª–∏ –æ–±—É—á–∞—é—â–∞—è –≤—ã–±–æ—Ä–∫–∞ —ç—Ç–æ –Ω–∞–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö, –Ω–∞ –∫–æ—Ç–æ—Ä—ã—Ö –º—ã –±—É–¥–µ–º –æ–±—É—á–∞—Ç—å –Ω–∞—à—É –º–æ–¥–µ–ª—å.

    VALID - –≤–∞–ª–∏–¥–∞—Ü–∏—è –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, –∫–∞–∫ –ø–æ–≤–µ–¥—ë—Ç —Å–µ–±—è –º–æ–¥–µ–ª—å —Å –Ω–µ–∑–Ω–∞–∫–æ–º—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏, –ø–æ—Å–ª–µ —Ç–æ–≥–æ –∫–∞–∫ –º–æ–¥–µ–ª—å –±—ã–ª–∞ –æ–±—É—á–µ–Ω–∞ –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ. –ï—Å–ª–∏ –ø—Ä–∞–≤–∏–ª—å–Ω–æ—Å—Ç—å –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –±—É–¥–µ—Ç –ø–∞–¥–∞—Ç—å, —Ç–æ –∑–Ω–∞—á–∏—Ç –º–æ–¥–µ–ª—å –ø–µ—Ä–µ–æ–±—É—á–µ–Ω–∞.

    TEST - c –ø–æ–º–æ—â—å—é —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏ –º—ã –ø—Ä–æ–≤–µ—Ä–∏–º, –∫–∞–∫ –º–æ–¥–µ–ª—å —Å–ø—Ä–∞–≤–ª—è–µ—Ç—Å—è —Å –Ω–æ–≤—ã–º–∏ –¥–∞–Ω–Ω—ã–º–∏ –∏ –Ω–µ —Å—Ç–æ–ª–∫–Ω—É–ª–∏—Å—å –ª–∏ –º—ã —Å –ø–µ—Ä–µ–æ–±—É—á–µ–Ω–∏–µ–º –º–æ–¥–µ–ª–∏. –¢–µ—Å—Ç–æ–≤–∞—è –≤—ã–±–æ—Ä–∫–∞ –Ω—É–∂–Ω–∞, —á—Ç–æ–±—ã –ø—Ä–∞–≤–∏–ª—å–Ω–æ –æ—Ü–µ–Ω–∏—Ç—å –≥–æ—Ç–æ–≤—É—é –º–æ–¥–µ–ª—å, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–æ—à–ª–∞ —ç—Ç–∞–ø—ã –æ–±—É—á–µ–Ω–∏—è –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–∏

    

```python
# –û–±—ä–µ–∫—Å—Ç–∞–º–∏ –Ω–∞–±–±–ª—é–¥–µ–Ω–∏–π (—Ç.–µ. –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏) –±—É–¥—É—Ç –≤—Å–µ —Å–µ—Ä–∏–∏ –∫—Ä–æ–º–µ is_ultra.
# –ò–º–µ–Ω–Ω–æ –Ω–∞ –æ—Å–Ω–æ–≤–µ —ç—Ç–∏—Ö –¥–∞–Ω–Ω—ã—Ö –Ω–∞—à–∞ –º–æ–¥–µ–ª—å –±—É–¥–µ—Ç —É—á–∏—Ç—å—Å—è –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞—Ç—å —Ç–∞—Ä–∏—Ñ
#   calls
#   minutes
#   messages
#   mb_used
features = data.drop(['is_ultra'], axis=1)

# –ü—Ä–∏–∑–Ω–∞–∫, –∫–æ—Ç–æ—Ä—ã–π –Ω–∞–º –Ω—É–∂–Ω–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞—Ç—å –æ–ø—Ä–µ–¥–µ–ª–∏–º –∫–∞–∫ —Ü–µ–ª–µ–≤–æ–π
target = data['is_ultra']
```

```python
# –¢–∞–∫ –∫–∞–∫ —Ç–µ—Å—Ç–æ–≤—ã–π –Ω–∞–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö –Ω–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω, —Ç–æ –¥–∞–Ω–Ω—ã–µ –Ω—É–∂–Ω–æ —Ä–∞–∑–±–∏—Ç—å –Ω–∞ —Ç—Ä–∏ —á–∞—Å—Ç–∏: 
# –æ–±—É—á–∞—é—â—É—é, –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é. 
# –†–∞–∑–º–µ—Ä—ã —Ç–µ—Å—Ç–æ–≤–æ–≥–æ –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–≥–æ –Ω–∞–±–æ—Ä–æ–≤ —É—Å—Ç–∞–Ω–æ–≤–∏–º —Ä–∞–≤–Ω—ã–º–∏ –¥—Ä—É–≥ –¥—Ä—É–≥—É, –æ—Å—Ç–∞–ª—å–Ω–æ–µ –æ—Ç–≤–µ–¥–µ–º –ø–æ–¥ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã–π –Ω–∞–±–æ—Ä.
# –†–∞–∑–æ–±—å–µ–º –∏—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –≤ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–∏ 3:1:1(60%:20%:20%). 

# —Ñ—É–Ω–∫—Ü–∏—è train_test_split —Ä–∞–∑–±–∏–≤–∞–µ—Ç –¥–∞—Ç–∞—Å–µ—Ç –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏, 
# –Ω–æ –º—ã –ø—Ä–∏–º–µ–Ω–∏–º —ç—Ç—É —Ñ—É–Ω–∫—Ü–∏—é, –¥–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã —Å–ø–µ—Ä–≤–∞ –ø–æ–ª—É—á–∏—Ç—å –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—ã–π –∏ –æ–±—É—á–∞—é—â–∏–π –Ω–∞–±–æ—Ä—ã, 
# –∞ –∑–∞—Ç–µ–º –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—ã–π(40%) –ø–æ–¥–µ–ª–∏–º –ø–æ–ø–æ–ª–∞–º –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—ã–π –∏ —Ç–µ—Å—Ç–æ–≤—ã–π –Ω–∞–±–æ—Ä—ã, —Ç–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º –∏—Ö –¥–æ–ª—è
# —Å–æ—Å—Ç–∞–≤–∏—Ç –ø–æ 20% –∫–∞–∂–¥–∞—è –æ—Ç –æ–±—â–µ–≥–æ –Ω–∞–±–æ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö, —á—Ç–æ —É—Å—Ç—Ä–∞–∏–≤–∞–µ—Ç –Ω–∞—à–µ —É—Å–ª–æ–≤–∏–µ.

# 1. –†–∞–∑–¥–µ–ª–∏–º –¥–∞—Ç–∞—Ñ—Ä—ç–π–º –Ω–∞ –¥–≤–µ –≤—ã–±–æ—Ä–∫–∏ –∏ –æ—Ç–≤–µ–¥–µ–º –∫–∞–∂–¥–æ–π –ø—Ä–æ—Ü–µ–Ω—Ç –¥–∞–Ω–Ω—ã—Ö –æ—Ç –æ–±—â–µ–≥–æ –Ω–∞–±–æ—Ä–∞:
#    features_train, target_train - 60% –¥–∞–Ω–Ω—ã—Ö
#    features_valid, target_valid - 40% –¥–∞–Ω–Ω—ã—Ö
features_train, features_valid, target_train, target_valid = train_test_split(features, 
                                                                              target, 
                                                                              test_size=0.4, 
                                                                              random_state=12345)

# 2. –¢–µ–ø–µ—Ä—å —Ä–∞–∑–¥–µ–ª–∏–º –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –≤—ã–±–æ—Ä–∫—É —Ä–∞–∑–º–µ—Ä–æ–º 40% –¥–∞–Ω–Ω—ã—Ö –æ—Ç –æ–±—â–µ–≥–æ –¥–∞—Ç–∞—Ñ—Ä—ç–π–º–∞ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é:
#     features_valid, target_valid - 50% –æ—Ç –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏ 40%, –ø–æ —Å—É—Ç–∏ —ç—Ç–æ 20% –ø—Ä–æ—Ü–µ–Ω—Ç–æ–≤ –¥–∞–Ω–Ω—ã—Ö –æ—Ç –æ–±—â–µ–≥–æ –¥–∞—Ç–∞—Ñ—Ä—ç–π–º–∞
#     features_test, target_test   - 50% –æ—Ç –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏ 40%, –ø–æ —Å—É—Ç–∏ —ç—Ç–æ 20% –ø—Ä–æ—Ü–µ–Ω—Ç–æ–≤ –¥–∞–Ω–Ω—ã—Ö –æ—Ç –æ–±—â–µ–≥–æ –¥–∞—Ç–∞—Ñ—Ä—ç–π–º–∞
features_valid, features_test, target_valid, target_test = train_test_split(features_valid,
                                                                            target_valid,
                                                                            test_size=0.5, 
                                                                            random_state=12345)
```

## –í—ã–±–æ—Ä –º–æ–¥–µ–ª–∏

–ò—Å—Å–ª–µ–¥—É–µ–º –º–æ–¥–µ–ª–∏ –∏ –æ–ø—Ä–µ–¥–µ–ª–∏–º –ª—É—á—à—É—é, —Å—Ä–µ–¥–∏ —Å–ª–µ–¥—É—é—â–∏—Ö:
- –¥–µ—Ä–µ–≤–æ —Ä–µ—à–µ–Ω–∏–π;
- —Å–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å;
- –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è;
- 
–û—Ç–º–µ—Ç–∏–º –¥–æ–ø–æ–ª–Ω–∏—Ç–µ—å–Ω–æ, —Ç–∞–∫ –∫–∞–∫ —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ —É –Ω–∞—Å —è–≤–Ω–æ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–π —Ç–æ –º—ã –±—É–¥–µ–º –∏—Å—Å–ª–µ–¥–æ–≤–∞—Ç—å –º–æ–¥–µ–ªb —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º –∞–ª–≥–æ—Ä–∏—Ç–º–æ–≤ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏,

–†–µ–∑—É–ª—å—Ç–∞—Ç —Ä–∞–±–æ—Ç—ã –º–æ–¥–µ–ª–∏ –±—É–¥–µ–º –æ—Ü–µ–Ω–∏–≤–∞—Ç—å –ø–æ –∫—Ä–∏—Ç–µ—Ä–∏—é –ø—Ä–∞–≤–∏–ª—å–Ω–æ—Å—Ç–∏(accuracy), —Ç.–µ. –æ—Ç–Ω–æ—à–µ–Ω–∏—é —á–∏—Å–ª–∞ –ø—Ä–∞–≤–∏–ª—å–Ω—ã—Ö –æ—Ç–≤–µ—Ç–æ–≤ –∫ —Ä–∞–∑–º–µ—Ä—É —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏. 

–¢–∞–∫ –∂–µ, –¥–ª—è –Ω–∞–∏–ª—É—á—à–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è accuracy –±—É–¥–µ–º —Ä–∞—Å—á–∏—Ç—ã–≤–∞—Ç—å –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –º–µ—Ç—Ä–∏–∫–∏:

    - recall - –ø–æ–ª–Ω–æ—Ç–∞. –ü–æ–∫–∞–∑—ã–≤–∞–µ—Ç –¥–æ–ª—é —Ç–∞—Ä–∏—Ñ–æ–≤ ¬´–£–ª—å—Ç—Ä–∞¬ª ‚Äî 1 —É–¥–∞–ª–æ—Å—å –æ–±–Ω–∞—Ä—É–∂–∏—Ç—å
    - precission - —Ç–æ—á–Ω–æ—Å—Ç—å. –ü–æ–∫–∞–∑—ã–≤–∞–µ—Ç, –∫–∞–∫–∞—è –¥–æ–ª—è –æ–±—ä–µ–∫—Ç–æ–≤, –æ—Ç–º–µ—á–µ–Ω–Ω—ã—Ö –º–æ–¥–µ–ª—å—é –∫–∞–∫ —Ç–∞—Ä–∏—Ñ ¬´–£–ª—å—Ç—Ä–∞¬ª ‚Äî 1 –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ —è–≤–ª—è—é—Ç—Å—è —Ç–∞—Ä–∏—Ñ–æ–º –£–ª—å—Ç—Ä–∞

```python
# –°–æ–∑–¥–∞–¥–∏–º –¥–∞—Ç–∞—Ñ—Ä—ç–π–º –≤ –∫–æ—Ç–æ—Ä—ã–π –±—É–¥–µ–º –∑–∞–ø–∏—Å—ã–≤–∞—Ç—å –±–µ–Ω—á–º–∞—Ä–∫–∏ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–π –¥–ª—è —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ –≤—ã–≤–æ–¥–∞
models_comparision_df = pd.DataFrame( columns=['model_name',
                                               'model_settings', 
                                               'accuracy',
                                               'recall',
                                               'precission',
                                               'elapsed_time',
                                               'errors'])
```

### –î–µ—Ä–µ–≤–æ —Ä–µ—à–µ–Ω–∏–π

```python
#–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å
best_model = None
#–õ—É—á—à–∞—è –≥–ª—É–±–∏–Ω–∞
best_depth = 0
# –õ—É—á—à–µ–µ –∫–∞—á–µ—Å—Ç–≤–æ
best_accuracy = 0
#–í—Ä–µ–º—è —Ä–∞–±–æ—Ç—ã –∞–ª–≥–æ—Ä–∏—Ç–º–∞
elapsed_time = 0
#–ö–æ–ª–ª–∏—á–µ—Å—Ç–≤–æ –æ—à–∏–±–æ–∫
errors_count = 0

# –ó–º–µ—Ä–∏–º –Ω–∞—á–∞–ª—å–Ω–æ–µ –≤—Ä–µ–º—è —Ä–∞–±–æ—Ç—ã –∞–ª–≥–æ—Ä–∏—Ç–º–∞
start_time = time.time()

for depth in range(1, 20):
    model = DecisionTreeClassifier(random_state=1234, max_depth=depth)
    
    # –æ–±—É—á–∏–º –º–æ–¥–µ–ª—å –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ
    model.fit(features_train, target_train)
    
    # –ø–æ–ª—É—á–∏–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –º–æ–¥–µ–ª–∏ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ
    predictions_valid = model.predict(features_valid)    
    
    # –†–∞—Å—á–∏—Ç–∞–µ–º –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å –ø—Ä–∞–≤–∏–ª—å–Ω–æ—Å—Ç–∏
    accuracy = accuracy_score(target_valid, predictions_valid)
     
    if accuracy > best_accuracy:       
        best_model = model
        best_depth = depth
        best_accuracy = accuracy
        # –†–∞—Å—á–∏—Ç–∞–µ–º –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å —Ç–æ—á–Ω–æ—Å—Ç–∏
        precission = precision_score(target_valid, predictions_valid)
        # –†–∞—Å—á–∏—Ç–∞–µ–º –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å –ø–æ–ª–Ω–æ—Å—Ç–∏
        recall = recall_score(target_valid, predictions_valid)
        # –†–∞—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–ª–∏—á–µ—Å—Ç–≤–æ –æ—à–∏–±–æ–∫
        errors_count = error_count(target_valid.tolist(), predictions_valid)
       
# –ó–º–µ—Ä–∏–º –∫–æ–Ω–µ—á–Ω–æ–µ –≤—Ä–µ–º—è —Ä–∞–±–æ—Ç—ã –∞–ª–≥–æ—Ä–∏—Ç–º–∞
finish_time = time.time()

# –í—ã—á–∏—Å–ª—è–µ–º –≤—Ä–µ–º—è –∑–∞–Ω—è—Ç–æ–µ –∞–ª–≥–æ—Ä–∏—Ç–º–æ–º
elapsed_time = finish_time - start_time

# –ó–∞–ø–∏—à–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
results_df = {'model_name': '–î–µ—Ä–µ–≤–æ —Ä–µ—à–µ–Ω–∏–π',
              'model_settings': best_model, 
              'accuracy': round(best_accuracy,4), 
              'recall': round(recall,4),
              'precission': round(precission,4), 
              'elapsed_time': round(elapsed_time, 2),
              'errors':errors_count}

#models_comparision_df = pd.append(results_df, ignore_index = True)
models_comparision_df = pd.concat([models_comparision_df, pd.DataFrame([results_df])], ignore_index=True)

# –í—ã–≤–∏–¥–µ–º —Ä–∞–∑–≤–µ—Ä–Ω—É—Ç—ã–π –ø—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω—ã–π –∞–Ω–∞–ª–∏–∑
print(f'–ù–∞—Å—Ç—Ä–æ–π–∫–∏ –º–æ–¥–µ–ª–∏ –ø–æ–±–µ–¥–∏—Ç–µ–ª—è: {best_model}')
print(f'–õ—É—á—à–∞—è –≥–ª—É–±–∏–Ω–∞: {best_depth}')
print(f'–õ—É—á—à–∞—è –ø—Ä–∞–≤–∏–ª—å–Ω–æ—Å—Ç—å: {round(best_accuracy,4)}')
print(f'   –≤ —Ç.—á. —Ç–æ—á–Ω–æ—Å—Ç—å: {round(precission,4)}')
print(f'   –≤ —Ç.—á. –ø–æ–ª–Ω–æ—Ç–∞: {round(recall,4)}')
print(f'–í—Ä–µ–º—è —Ä–∞–±–æ—Ç—ã –∞–ª–≥–æ—Ä–∏—Ç–º–∞: {round(elapsed_time, 2)} —Å–µ–∫.')
print("–û—à–∏–±–æ–∫:", errors_count)
```

### –°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å

#### –û–ø—Ü–∏—è 1. –ë–µ–∑ –ø–æ–¥–±–æ—Ä–∞ —á–∏—Å–ª–∞ –æ—Ü–µ–Ω—â–∏–∫–æ–≤ (estimators)

```python
#–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å
best_model = None
#–õ—É—á—à–∞—è –≥–ª—É–±–∏–Ω–∞
best_depth = 0
# –õ—É—á—à–µ–µ –∫–∞—á–µ—Å—Ç–≤–æ
best_accuracy = 0
#–í—Ä–µ–º—è —Ä–∞–±–æ—Ç—ã –∞–ª–≥–æ—Ä–∏—Ç–º–∞
elapsed_time = 0
#–ö–æ–ª–ª–∏—á–µ—Å—Ç–≤–æ –æ—à–∏–±–æ–∫
errors_count = 0

# –ó–º–µ—Ä–∏–º –Ω–∞—á–∞–ª—å–Ω–æ–µ –≤—Ä–µ–º—è —Ä–∞–±–æ—Ç—ã –∞–ª–≥–æ—Ä–∏—Ç–º–∞
start_time = time.time()

for depth in range(1, 20):
    model = RandomForestClassifier(random_state=1234, max_depth=depth)
    # –æ–±—É—á–∏–º –º–æ–¥–µ–ª—å –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ
    model.fit(features_train, target_train)
    
    # –ø–æ–ª—É—á–∏–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –º–æ–¥–µ–ª–∏ –Ω–∞ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ
    predictions_valid = model.predict(features_valid)
    
    accuracy = accuracy_score(target_valid, predictions_valid)
    #accuracy = model.score(features_valid, target_valid)
    if accuracy > best_accuracy:       
        best_model = model
        best_depth = depth
        best_accuracy = accuracy
        # –†–∞—Å—á–∏—Ç–∞–µ–º –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å —Ç–æ—á–Ω–æ—Å—Ç–∏
        precission = precision_score(target_valid, predictions_valid)
        # –†–∞—Å—á–∏—Ç–∞–µ–º –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å –ø–æ–ª–Ω–æ—Å—Ç–∏
        recall = recall_score(target_valid, predictions_valid)
        # –†–∞—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–ª–∏—á–µ—Å—Ç–≤–æ –æ—à–∏–±–æ–∫
        errors_count = error_count(target_valid.tolist(), predictions_valid)

# –ó–º–µ—Ä–∏–º –∫–æ–Ω–µ—á–Ω–æ–µ –≤—Ä–µ–º—è —Ä–∞–±–æ—Ç—ã –∞–ª–≥–æ—Ä–∏—Ç–º–∞
finish_time = time.time()

# –í—ã—á–∏—Å–ª—è–µ–º –≤—Ä–µ–º—è –∑–∞–Ω—è—Ç–æ–µ –∞–ª–≥–æ—Ä–∏—Ç–º–æ–º
elapsed_time = finish_time - start_time

# –ó–∞–ø–∏—à–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
results_df = {'model_name': '–°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å. –û–ø.1',
              'model_settings': best_model, 
              'accuracy': round(best_accuracy,4), 
              'recall': round(recall,4),
              'precission': round(precission,4), 
              'elapsed_time': round(elapsed_time, 2),
              'errors':errors_count}

#models_comparision_df = models_comparision_df.append(results_df, ignore_index = True)
models_comparision_df = pd.concat([models_comparision_df, pd.DataFrame([results_df])], ignore_index=True)

print(f'–ù–∞—Å—Ç—Ä–æ–π–∫–∏ –º–æ–¥–µ–ª–∏ –ø–æ–±–µ–¥–∏—Ç–µ–ª—è: {best_model}')
print(f'–õ—É—á—à–∞—è –≥–ª—É–±–∏–Ω–∞: {best_depth}')
print(f'–õ—É—á—à–∞—è –ø—Ä–∞–≤–∏–ª—å–Ω–æ—Å—Ç—å: {round(best_accuracy,4)}')
print(f'   –≤ —Ç.—á. —Ç–æ—á–Ω–æ—Å—Ç—å: {round(precission,4)}')
print(f'   –≤ —Ç.—á. –ø–æ–ª–Ω–æ—Ç–∞: {round(recall,4)}')
print(f'–í—Ä–µ–º—è —Ä–∞–±–æ—Ç—ã –∞–ª–≥–æ—Ä–∏—Ç–º–∞: {round(elapsed_time, 2)} —Å–µ–∫.')
print("–û—à–∏–±–æ–∫:", errors_count)
```

```python
# –°–æ–∑–¥–∞–¥–∏–º –¥–∞—Ç–∞—Ñ—Ä—ç–π–º –≤ –∫–æ—Ç–æ—Ä—ã–π –±—É–¥–µ–º –∑–∞–ø–∏—Å—ã–≤–∞—Ç—å –±–µ–Ω—á–º–∞—Ä–∫–∏ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–π –¥–ª—è —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ –≤—ã–≤–æ–¥–∞
models_comparision_test_df = pd.DataFrame( columns=['model_name',
                                                   'model_settings', 
                                                   'accuracy',
                                                   'recall',
                                                   'precission',
                                                   'elapsed_time',
                                                   'errors'])
```

### –û–ø—Ü–∏—è 1. –ë–µ–∑ –ø–æ–¥–±–æ—Ä–∞ —á–∏—Å–ª–∞ –æ—Ü–µ–Ω—â–∏–∫–æ–≤ (estimators)

```python
#–õ—É—á—à–∞—è –≥–ª—É–±–∏–Ω–∞
depth = 8

#
errors_count = 0

#–í—Ä–µ–º—è —Ä–∞–±–æ—Ç—ã –∞–ª–≥–æ—Ä–∏—Ç–º–∞
elapsed_time = 0

# –ó–º–µ—Ä–∏–º –Ω–∞—á–∞–ª—å–Ω–æ–µ –≤—Ä–µ–º—è —Ä–∞–±–æ—Ç—ã –∞–ª–≥–æ—Ä–∏—Ç–º–∞
start_time = time.time()

model = RandomForestClassifier(random_state=1234, max_depth=depth)
# –æ–±—É—á–∏–º –º–æ–¥–µ–ª—å –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ
model.fit(features_train, target_train)
    
# –ø–æ–ª—É—á–∏–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
predictions_test = model.predict(features_test)

# –†–∞—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–ª–∏—á–µ—Å—Ç–≤–æ –æ—à–∏–±–æ–∫
errors_count = error_count(target_test.tolist(), predictions_test)

# –†–∞—Å—á–∏—Ç–∞–µ–º –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å –ø—Ä–∞–≤–∏–ª—å–Ω–æ—Å—Ç–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
accuracy = accuracy_score(target_test, predictions_test)
# –†–∞—Å—á–∏—Ç–∞–µ–º –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å —Ç–æ—á–Ω–æ—Å—Ç–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
precission = precision_score(target_test, predictions_test)
# –†–∞—Å—á–∏—Ç–∞–µ–º –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å –ø–æ–ª–Ω–æ—Å—Ç–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
recall = recall_score(target_test, predictions_test)

# –ó–º–µ—Ä–∏–º –∫–æ–Ω–µ—á–Ω–æ–µ –≤—Ä–µ–º—è —Ä–∞–±–æ—Ç—ã –∞–ª–≥–æ—Ä–∏—Ç–º–∞
finish_time = time.time()

# –í—ã—á–∏—Å–ª—è–µ–º –≤—Ä–µ–º—è –∑–∞–Ω—è—Ç–æ–µ –∞–ª–≥–æ—Ä–∏—Ç–º–æ–º
elapsed_time = finish_time - start_time

# –ó–∞–ø–∏—à–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
results_df = {'model_name': '–°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å. –û–ø.1',
              'model_settings': model, 
              'accuracy': round(accuracy,4), 
              'recall': round(recall,4),
              'precission': round(precission,4), 
              'elapsed_time': round(elapsed_time, 2),
              'errors': errors_count}

#models_comparision_test_df = models_comparision_test_df.append(results_df, ignore_index = True)
models_comparision_df = pd.concat([models_comparision_df, pd.DataFrame([results_df])], ignore_index=True)


print(f'–ù–∞—Å—Ç—Ä–æ–π–∫–∏ –º–æ–¥–µ–ª–∏: {model}')
print(f'–ì–ª—É–±–∏–Ω–∞: {depth}')
print(f'–ü—Ä–∞–≤–∏–ª—å–Ω–æ—Å—Ç—å: {round(accuracy,4)}')
print(f'   –≤ —Ç.—á. —Ç–æ—á–Ω–æ—Å—Ç—å: {round(precission,4)}')
print(f'   –≤ —Ç.—á. –ø–æ–ª–Ω–æ—Ç–∞: {round(recall,4)}')
print(f'–í—Ä–µ–º—è —Ä–∞–±–æ—Ç—ã –∞–ª–≥–æ—Ä–∏—Ç–º–∞: {round(elapsed_time, 2)} —Å–µ–∫.')
print("–û—à–∏–±–æ–∫:", errors_count)
```

```python
#–õ—É—á—à–∞—è –≥–ª—É–±–∏–Ω–∞
depth = 9
# –õ—É—á—à–µ–µ –∫–∞—á–µ—Å—Ç–≤–æ
accuracy = 0
#–í—Ä–µ–º—è —Ä–∞–±–æ—Ç—ã –∞–ª–≥–æ—Ä–∏—Ç–º–∞
elapsed_time = 0

#
errors_count = 0

# –ó–º–µ—Ä–∏–º –Ω–∞—á–∞–ª—å–Ω–æ–µ –≤—Ä–µ–º—è —Ä–∞–±–æ—Ç—ã –∞–ª–≥–æ—Ä–∏—Ç–º–∞
start_time = time.time()


model = DecisionTreeClassifier(random_state=1234, max_depth=depth)
    
# –æ–±—É—á–∏–º –º–æ–¥–µ–ª—å –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ
model.fit(features_train, target_train)
    
# –ø–æ–ª—É—á–∏–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
predictions_test = model.predict(features_test)
 
# –†–∞—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–ª–∏—á–µ—Å—Ç–≤–æ –æ—à–∏–±–æ–∫
errors_count = error_count(target_test.tolist(), predictions_test)

# –†–∞—Å—á–∏—Ç–∞–µ–º –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å –ø—Ä–∞–≤–∏–ª—å–Ω–æ—Å—Ç–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
accuracy = accuracy_score(target_test, predictions_test)
# –†–∞—Å—á–∏—Ç–∞–µ–º –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å —Ç–æ—á–Ω–æ—Å—Ç–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
precission = precision_score(target_test, predictions_test)
# –†–∞—Å—á–∏—Ç–∞–µ–º –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å –ø–æ–ª–Ω–æ—Å—Ç–∏ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
recall = recall_score(target_test, predictions_test)

# –ó–º–µ—Ä–∏–º –∫–æ–Ω–µ—á–Ω–æ–µ –≤—Ä–µ–º—è —Ä–∞–±–æ—Ç—ã –∞–ª–≥–æ—Ä–∏—Ç–º–∞
finish_time = time.time()

# –í—ã—á–∏—Å–ª—è–µ–º –≤—Ä–µ–º—è –∑–∞–Ω—è—Ç–æ–µ –∞–ª–≥–æ—Ä–∏—Ç–º–æ–º
elapsed_time = finish_time - start_time

# –ó–∞–ø–∏—à–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
results_df = {'model_name': '–î–µ—Ä–µ–≤–æ —Ä–µ—à–µ–Ω–∏–π',
              'model_settings': model, 
              'accuracy': round(accuracy,4), 
              'recall': round(recall,4),
              'precission': round(precission,4), 
              'elapsed_time': round(elapsed_time, 2),
              'errors':errors_count}

#models_comparision_test_df = models_comparision_test_df.append(results_df, ignore_index = True)
models_comparision_df = pd.concat([models_comparision_df, pd.DataFrame([results_df])], ignore_index=True)


# –í—ã–≤–∏–¥–µ–º —Ä–∞–∑–≤–µ—Ä–Ω—É—Ç—ã–π –ø—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω—ã–π –∞–Ω–∞–ª–∏–∑
print(f'–ù–∞—Å—Ç—Ä–æ–π–∫–∏ –º–æ–¥–µ–ª–∏: {model}')
print(f'–ì–ª—É–±–∏–Ω–∞: {depth}')
print(f'–ü—Ä–∞–≤–∏–ª—å–Ω–æ—Å—Ç—å: {round(accuracy,4)}')
print(f'   –≤ —Ç.—á. —Ç–æ—á–Ω–æ—Å—Ç—å: {round(precission,4)}')
print(f'   –≤ —Ç.—á. –ø–æ–ª–Ω–æ—Ç–∞: {round(recall,4)}')
print(f'–í—Ä–µ–º—è —Ä–∞–±–æ—Ç—ã –∞–ª–≥–æ—Ä–∏—Ç–º–∞: {round(elapsed_time, 2)} —Å–µ–∫.')
print("–û—à–∏–±–æ–∫:", errors_count)
```
